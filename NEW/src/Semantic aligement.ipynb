```python
import pandas as pd
from openai import OpenAI
import json
import time

# ===========================================================
# 1. ì„¤ì • (Configuration)
# ===========================================================
API_KEY = "-"  #  ì—¬ê¸°ì— ë³¸ì¸ì˜ OpenAI API Keyë¥¼ ì…ë ¥í•˜ì„¸ìš”!
client = OpenAI(api_key=API_KEY)

INPUT_FILE = "evaluation_set_v1.csv"            # ì•„ê¹Œ ë§Œë“  í‰ê°€ì…‹
OUTPUT_FILE = "evaluation_result_semantic_final.xlsx" # ìµœì¢… ê²°ê³¼ íŒŒì¼

# ===========================================================
# 2. í”„ë¡¬í”„íŠ¸ ì„¤ê³„ (ìµœì¢… í™•ì • ìŠ¤í‚¤ë§ˆ ë°˜ì˜)
# ===========================================================

# [DJ ì—­í• ] ë°±ì—”ë“œì— ì „ë‹¬í•œ ìµœì¢… ìŠ¤í‚¤ë§ˆ ê·¸ëŒ€ë¡œ ì ìš©
SYSTEM_PROMPT_DJ = """
# Role
ë‹¹ì‹ ì€ ì‚¬ìš©ìì˜ ìƒí™©(Context)ê³¼ ì·¨í–¥(Preference)ì— ë§ì¶°, Spotifyì— ì‹¤ì¡´í•˜ëŠ” ìŒì•…ì„ ì¶”ì²œí•˜ëŠ” AI DJì…ë‹ˆë‹¤.

# Critical Constraints
1. **Real Song Only**: ë°˜ë“œì‹œ ì‹¤ì œë¡œ ì¡´ì¬í•˜ëŠ” ê³¡ì„ ì¶”ì²œí•´ì•¼ í•©ë‹ˆë‹¤. (í• ë£¨ì‹œë„¤ì´ì…˜ ê¸ˆì§€)
2. **Context Alignment**: ëª©í‘œ(Goal)ì™€ ì†ŒìŒ(Decibel)ì„ ìµœìš°ì„ ìœ¼ë¡œ ê³ ë ¤í•˜ì„¸ìš”.
   - ì˜ˆ: Sleep ëª©í‘œ -> ë½/ëŒ„ìŠ¤ ê¸ˆì§€.
3. **Format**: ë°˜ë“œì‹œ ì•„ë˜ JSON í¬ë§·ì„ ì§€í‚¤ì„¸ìš”.

# Output JSON Structure
{
  "recommendation_meta": {
    "reasoning": "ì¶”ì²œ ì´ìœ  (í•œê¸€ 1ë¬¸ì¥)",
    "primary_tag": "mood_tag"
  },
  "track_info": {
    "artist_name": "Exact Artist Name",
    "track_id": "Exact Song Title Spotify ID",
    "track_title": "Exact Song Title"
  },
  "target_audio_features": {
    "min_tempo": int,
    "max_tempo": int,
    "target_energy": (float 0.0~1.0),
    "target_instrumentalness": (float 0.0~1.0)
    "target_valence": (float 0.0~1.0),
    "target_acousticness": (float 0.0~1.0)
  }
}
"""

# [Judge ì—­í• ] ë…¸ë˜ ì œëª©ê³¼ ì´ìœ ë¥¼ ë³´ê³  ì í•©ì„± í‰ê°€
SYSTEM_PROMPT_JUDGE = """
# Role
ë‹¹ì‹ ì€ ìŒì•… í‰ë¡ ê°€ì´ì QA Judgeì…ë‹ˆë‹¤.
AIê°€ ì¶”ì²œí•œ 'ë…¸ë˜(Song)'ì™€ 'ì¶”ì²œ ì‚¬ìœ (Reasoning)'ê°€ ì£¼ì–´ì§„ 'ìƒí™©(Context)'ì— ì í•©í•œì§€ í‰ê°€í•©ë‹ˆë‹¤.

# Task
1. ì¶”ì²œëœ ë…¸ë˜(Artist - Title)ê°€ ì‹¤ì œë¡œ ì¡´ì¬í•˜ëŠ”ì§€, ì–´ë–¤ ë¶„ìœ„ê¸°ì¸ì§€ ë‹¹ì‹ ì˜ ì§€ì‹(Knowledge Base)ì„ í†µí•´ íŒŒì•…í•˜ì„¸ìš”.
2. ì¶”ì²œ ì‚¬ìœ (Reasoning)ê°€ ìƒí™©ì— ë§ê²Œ ë…¼ë¦¬ì ì¸ì§€ í™•ì¸í•˜ì„¸ìš”.

# Scoring Criteria (1-5)
- 5ì  (Perfect): ìƒí™©ì— ì™„ë²½í•˜ê²Œ ì–´ìš¸ë¦¬ëŠ” ëª…ê³¡ì´ë©°, ì¶”ì²œ ì‚¬ìœ ë„ ë…¼ë¦¬ì ì„.
- 3ì  (Fair): ë‚˜ì˜ì§€ ì•Šìœ¼ë‚˜, ë² ìŠ¤íŠ¸ ì´ˆì´ìŠ¤ëŠ” ì•„ë‹˜.
- 1ì  (Fail): ìƒí™©ê³¼ ì •ë°˜ëŒ€(ë¶„ìœ„ê¸° íŒŒê´´) ë˜ëŠ” ì¡´ì¬í•˜ì§€ ì•ŠëŠ” ë…¸ë˜(Hallucination).

# Output JSON
{"score": int, "reason": "string"}
"""

# ===========================================================
# 3. í•¨ìˆ˜ ì •ì˜
# ===========================================================

def get_song_recommendation(row):
    """ [Step 1] GPTì—ê²Œ ìµœì¢… ìŠ¤í‚¤ë§ˆë¡œ ì¶”ì²œ ìš”ì²­ """
    user_msg = f"""
    Location: {row['Location']}
    Decibel: {row['Decibel']}
    Goal: {row['Goal']}
    Preference: {row['User Pref (Genre/Artist)']}
    """
    try:
        response = client.chat.completions.create(
            model="gpt-4o",
            messages=[
                {"role": "system", "content": SYSTEM_PROMPT_DJ},
                {"role": "user", "content": user_msg}
            ],
            response_format={"type": "json_object"},
            temperature=0.7
        )
        return json.loads(response.choices[0].message.content)
    except Exception as e:
        print(f"âŒ Gen Error: {e}")
        return {}

def evaluate_semantic_alignment(row, artist, title, reasoning, target_features):
    """ [Step 2] LLM Judgeì—ê²Œ ì˜ë¯¸ì  ì¼ì¹˜ì„± í‰ê°€ ìš”ì²­ """
    judge_msg = f"""
    [User Context]
    Location: {row['Location']}, Goal: {row['Goal']}, Decibel: {row['Decibel']}
    User Pref: {row['User Pref (Genre/Artist)']}
    
    [AI Recommendation]
    Song: {artist} - {title}
    Reasoning: {reasoning}
    Target Features (AI Intention): {target_features}
    
    [Expected Vibe]
    {row['Expected Keywords / Vibe (ê¸°ëŒ€ ê²°ê³¼)']}
    
    ìœ„ ë…¸ë˜ì™€ ì¶”ì²œ ì‚¬ìœ ê°€ ì´ ìƒí™©ì— ì í•©í•œì§€ í‰ê°€í•´ì£¼ì„¸ìš”.
    """
    try:
        response = client.chat.completions.create(
            model="gpt-4o",
            messages=[
                {"role": "system", "content": SYSTEM_PROMPT_JUDGE},
                {"role": "user", "content": judge_msg}
            ],
            response_format={"type": "json_object"}
        )
        return json.loads(response.choices[0].message.content)
    except Exception as e:
        return {"score": 0, "reason": f"Error: {e}"}

# ===========================================================
# 4. ë©”ì¸ ì‹¤í–‰ (ìˆ˜ì •ë¨)
# ===========================================================
print("\nğŸš€ [Final Schema] ì˜ë¯¸ì  ì¼ì¹˜ì„± í‰ê°€ ì‹œì‘...")
df = pd.read_csv(INPUT_FILE)
results = []

for index, row in df.iterrows():
    print(f"â–¶ Case {index + 1}: {row['Location']} ({row['Goal']})...", end=" ")
    
    # 1. ì¶”ì²œ ìƒì„±
    gpt_output = get_song_recommendation(row)
    
    # ë°ì´í„° íŒŒì‹±
    meta = gpt_output.get('recommendation_meta', {})
    track = gpt_output.get('track_info', {})
    target = gpt_output.get('target_audio_features', {})
    
    artist = track.get('artist_name', "Unknown")
    title = track.get('track_title', "Unknown")
    reasoning = meta.get('reasoning', "No reasoning")
    # â–¼ [ì¶”ê°€ë¨] GPTê°€ ì„¤ì •í•œ ë¶„ìœ„ê¸° íƒœê·¸ ê°€ì ¸ì˜¤ê¸°
    primary_tag = meta.get('primary_tag', "Unknown") 
    
    print(f"ğŸµ {artist} - {title} (Tag: {primary_tag})")
    
    # 2. í’ˆì§ˆ í‰ê°€ (LLM Judge)
    judge_result = evaluate_semantic_alignment(row, artist, title, reasoning, target)
    
    # 3. ì €ì¥
    results.append({
        "Case ID": row['ID'],
        "Scenario": f"{row['Location']} / {row['Goal']}",
        "Conflict Type": row['Type'],
        
        # ì¶”ì²œ ê²°ê³¼
        "Recommended Artist": artist,
        "Recommended Title": title,
        "Primary Tag": primary_tag, # ğŸ‘ˆ ì—‘ì…€ì— ì €ì¥!
        "Reasoning": reasoning,
        
        # [ìˆ˜ì •ë¨] ë¶„ì„ìš© Target Features ì „ì²´ ì €ì¥ (GPTì˜ ì˜ë„)
        "Target Energy": target.get('target_energy', None),
        "Target Valence": target.get('target_valence', None),
        "Target Instrumentalness": target.get('target_instrumentalness', None),
        "Target Acousticness": target.get('target_acousticness', None),
        "Target Min Tempo": target.get('min_tempo', None),
        "Target Max Tempo": target.get('max_tempo', None),

        # í‰ê°€ ì ìˆ˜
        "LLM Score": judge_result['score'],
        "Judge Reason": judge_result['reason']
    })
    
    time.sleep(0.5)


# ===========================================================
# 5. ê²°ê³¼ ì €ì¥ ë° KPI í™•ì¸
# ===========================================================
result_df = pd.DataFrame(results)
result_df.to_excel(OUTPUT_FILE, index=False)

avg_score = result_df['LLM Score'].mean()
fatal_errors = len(result_df[result_df['LLM Score'] == 1])

print("\n" + "="*50)
print(f"âœ… í‰ê°€ ì™„ë£Œ! íŒŒì¼: {OUTPUT_FILE}")
print(f"ğŸ“Š [KPI 1] ì˜ë¯¸ì  ì¼ì¹˜ì„± í‰ê·  ì ìˆ˜: {avg_score:.2f} / 5.0 (ëª©í‘œ: 4.5)")
print(f"ğŸ›¡ï¸ [KPI 2] ì¹˜ëª…ì  ì˜¤ë¥˜(1ì ) ë°œìƒ: {fatal_errors}ê±´ (ëª©í‘œ: 0ê±´)")
print("="*50)
```
