### 데이터분석 파트 전체 파이프라인 다이어그램 요약

**1️ 문제 정의 & 목표 설정**

* 규칙 기반 추천 시스템 정의 (AI 미도입)
* KPI 설정: 커버리지 ≥90%, 적합도 ≥80%

**2️ 데이터 수집 & 구조화**

* 소스: 위치(GPS), 데시벨(센서/API), 목표(사용자 입력), Spotify(플레이리스트)
* 이상치 제거, 결측치 처리 → CSV/JSON 표준 포맷화

**3️ EDA (탐색적 데이터 분석)**

* 위치별 평균 데시벨, 목표별 BPM, 분위기 상관관계 시각화
* 규칙 설계 근거 확보

**4️ 규칙 정의 (Rule Mapping)**

* (장소 × 데시벨 × 목표) → (장르, 무드, BPM 범위, 보컬유무)
* Spotify 시드 키워드/플레이리스트 룰 작성
* 예시: 카페+70dB+집중 → Lo-Fi / 80~90BPM / Chill

**5️ 규칙 엔진 연동 (Core Integration)**

* JSON 기반 규칙 데이터 → 백엔드에 전달
* 입력검증 → 규칙매칭 → 결과 → Fallback (Neutral/Ambient)

**6️ 타당성 검증 & 피드백**

* 커버리지, Edge Case, Fallback 비율 측정
* 사용자 설문(5점 척도) 및 로그 분석
* 대시보드로 시각화 (히트맵, 산점도)

**7️ 규칙 튜닝 & 결과 보고**

* 파라미터 재조정 (데시벨 경계, BPM 가중치 등)
* 최종 산출물: JSON 규칙셋 + 분석 리포트 + 시각화 대시보드

---



> 위치·데시벨·목표 데이터를 기반으로 규칙을 설계하고,
> 이를 백엔드에서 직접 사용할 수 있는 구조로 전달해
> 추천 결과의 타당성을 데이터로 검증하는 전체 파이프라인.
